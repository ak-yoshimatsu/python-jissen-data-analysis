# matplotlib
## グラフ描画
pyplotモジュールを使ってグラフ描画

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([1, 2, 3, 4])  # plot: 線によるグラフ描画
plt.show()  # show関数：対話ウィンドウに表示
```

```python
# x軸, y軸
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
# 線以外のグラフ描画
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], "o")
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], "o")
plt.axis([0, 6, 0, 10])  # x軸を0~6, y軸を0~10表示させる
plt.show()
```

## 複数のグラフを1つの図に描画

```python
t = np.arange(0, 5, 0.2)
plt.plot(t, t, "r--", label="y=x")
plt.plot(t, t**2, "bs", label="y=x**2")
plt.plot(t, t**3, "g^", label="y=x**3")
plt.legend()  # 図の中にラベルを表示
plt.show()
```

## 散布図の作成
`scatter` 関数

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 100  # サイズ指定
colors = np.random.rand(50)  # カラー指定 Colormap ( https://matplotlib.org/stable/users/explain/customizing.html )
plt.scatter(x, y, s=sizes, c=colors)
plt.show()
```

## 棒グラフの作成

```python
values = [1, 2, 3, 4, 5, 6]
x_pos = np.arange(len(values))
object = ["a", "b", "c", "d", "e", "f"]

plt.bar(x=x_pos, height=values)
plt.xticks(ticks=x_pos, labels=object)  # x軸にラベル付け
plt.show()
```

## 円グラフ作成

```python
rates = [10, 20, 30, 40]
labels = ["Python", "C++", "Ruby", "Java"]
colors = ["red", "green", "yellow", "blue"]

plt.pie(
    x=rates, labels=labels, colors=colors, autopct="%1.1f%%"  # autopct: パーセント表示
)
plt.axis("equal")  # グラフ軸の属性を設定 "equal"を渡すことで正円になる
plt.show()
```
