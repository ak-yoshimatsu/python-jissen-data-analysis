```python
# pandasでcsv読み込み
import pandas as pd

df_sales_2021 = pd.read_csv("data/売上テーブル_2021.csv", encoding="SJIS")
df_sales_2021.head()
```

```python
df_sales_2022 = pd.read_csv("data/売上テーブル_2022.csv", encoding="SJIS")
df_sales_2022.head()
```

```python
df_customer = pd.read_csv("data/顧客テーブル.csv", encoding="SJIS")
df_customer.head()
```

## データの結合（ユニオン）: concat

```python
# ignore_index: 結合前のインデックスを無視して、結合後のインデックスを適用
df_sales = pd.concat([df_sales_2021, df_sales_2022], ignore_index=True)
print(df_sales.head())
print(df_sales.tail())
print(len(df_sales))
```

## データの結合（ジョイン）: merge

```python
df_sales_customer = pd.merge(left=df_sales, right=df_customer, how="left", on="顧客ID")
df_sales_customer.head()
```

## データクリーニング
### 欠損値の処理
欠損値の処理はだいたい「欠損値を含むレコードや項目を消す」「欠損値を別の値で置き換える」「何もしない」のどれか

```python
# 例：欠損値を除外 dropna, subset: 特定の項目のみに絞って判定, inplace: 上書き
df_sales_customer.dropna(subset=["売上"], inplace=True)
df_sales_customer.head()
df_sales_customer.loc[df_sales_customer["売上"].isnull()]
```

### 異常値の処理

```python
df_sales_customer.loc[(df_sales_customer["売上"] == 9999999) | (df_sales_customer["売上"] == 0)]
# 例：異常値は除外
df_sales_customer = df_sales_customer.loc[(df_sales_customer["売上"] != 9999999) & (df_sales_customer["売上"] != 0)]
df_sales_customer.head()
```

datetime型へ変換

```python
df_sales_customer['売上日'] = pd.to_datetime(df_sales_customer['売上日'])
df_sales_customer.head()
```

```python
df_sales_customer.loc[df_sales_customer['売上日'].dt.year == 2022]
```

曜日の追加

```python
df_sales_customer['売上日_曜日'] = df_sales_customer['売上日'].dt.day_name()
df_sales_customer.head()
```

`df.copy()` でコピーする。`=` をすると、リストと同じように参照先情報だけコピーされる

```python
df_sales_customer_2 = df_sales_customer.copy()
df_sales_customer_2.head()
```
