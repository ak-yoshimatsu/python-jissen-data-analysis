## ファイルアクセス

```python
# pandasでcsv読み込み
import pandas as pd

df_sales_2021 = pd.read_csv("data/売上テーブル_2021.csv", encoding="SJIS")
df_sales_2021.head()
```

```python
df_sales_2022 = pd.read_csv("data/売上テーブル_2022.csv", encoding="SJIS")
df_sales_2022.head()
```

```python
df_customer = pd.read_csv("data/顧客テーブル.csv", encoding="SJIS")
df_customer.head()
```

## データの結合（ユニオン）: concat

```python
# ignore_index: 結合前のインデックスを無視して、結合後のインデックスを適用
df_sales = pd.concat([df_sales_2021, df_sales_2022], ignore_index=True)
print(df_sales.head())
print(df_sales.tail())
print(len(df_sales))
```

## データの結合（ジョイン）: merge

```python
df_sales_customer = pd.merge(left=df_sales, right=df_customer, how="left", on="顧客ID")
df_sales_customer.head()
```

## データクリーニング
### 欠損値の処理
欠損値の処理はだいたい「欠損値を含むレコードや項目を消す」「欠損値を別の値で置き換える」「何もしない」のどれか

```python
# 例：欠損値を除外 dropna, subset: 特定の項目のみに絞って判定, inplace: 上書き
df_sales_customer.dropna(subset=["売上"], inplace=True)
df_sales_customer.head()
df_sales_customer.loc[df_sales_customer["売上"].isnull()]
```

### 異常値の処理

```python
df_sales_customer.loc[(df_sales_customer["売上"] == 9999999) | (df_sales_customer["売上"] == 0)]
# 例：異常値は除外
df_sales_customer = df_sales_customer.loc[(df_sales_customer["売上"] != 9999999) & (df_sales_customer["売上"] != 0)]
df_sales_customer.head()
```

datetime型へ変換

```python
df_sales_customer['売上日'] = pd.to_datetime(df_sales_customer['売上日'])
df_sales_customer.head()
```

```python
df_sales_customer.loc[df_sales_customer['売上日'].dt.year == 2022]
```

曜日の追加

```python
df_sales_customer['売上日_曜日'] = df_sales_customer['売上日'].dt.day_name()
df_sales_customer.head()
```

## データフレームのコピー
`df.copy()` でコピーする。`=` をすると、リストと同じように参照先情報だけコピーされる

```python
df_sales_customer_2 = df_sales_customer.copy()
df_sales_customer_2.head()
```

インデックスの初期化

```python
df_sales_customer_2.reset_index(drop=True, inplace=True)  # drop: 元々あったインデックスを項目に追加するかどうか
df_sales_customer_2.tail()
```

## 行（レコード）の操作
行方向の操作は基本的にPandasの機能を用いる。for文を用いない。

```python
# for文を用いる例
tokyo_sales = 0
for index, row in df_sales_customer.iterrows():
    if row['都道府県'] == "東京都":
        tokyo_sales += row['売上']
print(tokyo_sales)
```

```python
# Pandasの機能を用いる例
df_sales_customer.loc[df_sales_customer['都道府県']=='東京都', '売上'].sum()
```

## データの可視化

agg: 集計 集計したい項目と集計方法を指定する。
集計方法<br>
合計：sum、平均値：mean、中央値：median

```python
df_sales_customer.groupby('顧客区分').agg({'売上': 'sum'})
```

```python
import matplotlib.pyplot as plt
import japanize_matplotlib

viz = df_sales_customer.groupby("顧客区分").agg({"売上": "sum"})
plt.bar(viz.index, viz["売上"])  # plt.barは棒グラフ、plt.bar(x軸, y軸)が基本
```

## pivot_table（クロス集計表）

```python
df_sales_customer['年'] = df_sales_customer['売上日'].dt.year
df_sales_customer.pivot_table(index=['顧客区分'], columns=['年'], values='売上', aggfunc='sum')
```
